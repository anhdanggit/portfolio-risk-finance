# Example: MovieLens 1M Dataset
* This provides a number of collections of movie ratings data (from MovieLens in the late 1990s and early 2000s).
Including: movie ratings, movie metadata (genres and year), and demographic data about users (age, zip code, gender identification, and occupation). Such data is interesting for Recommendation Systems. 
* The data set contains 1 million ratings from 6,000 users and 4,000 movies

```python
import pandas as pd 
```

```python
# from jupyterthemes import jtplot
# jtplot.style(theme="chesterish")
```

```python
pd.options.display.max_rows = 10 # Make display smaller
## read user data
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('data/movielens/users.dat', sep='::', header=None, names=unames)
## read ratings data
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('data/movielens/ratings.dat', sep='::', header=None, names=rnames)
## read movie data
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('data/movielens/movies.dat', sep='::', header=None, names=mnames)
```

```python
## data
users.head()
```

```python
ratings.head()
```

```python
movies.head()
```

```python
## Merge the data together
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
```

## Pivot Summary the data

```python
## Take the mean rating for each film by gender
g = data.groupby(['title','gender'])
g.mean()['rating'].unstack()
```

```python
## or using the pivot table
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings.head()
```

```python
## filter the movies having at least 250 ratings
nratings_by_title = data.groupby('title').size()
nratings_by_title.head()
```

```python
active_titles = nratings_by_title.index[nratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
## take the top 10 highest rating among female
top_female_rating = mean_ratings.sort_values(by='F', ascending=False)
top_female_rating.head(10)
```

## Some manipulations

```python
## Measure the rating disagreement between male and female 
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F'] ## male rate much lower than F
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff.head(10)
```

```python
## Reverse other
sorted_by_diff[::-1].head(10) ## reverse to get top 10 preffer by man by women don't like
```

```python
## Most disagree among users regardless genders (measured by variance and std)
ratings_std_by_title = data.groupby('title')['rating'].std()
ratings_std_by_title = ratings_std_by_title.loc[active_titles] ## only titles with > 250 ratings
ratings_std_by_title.sort_values(ascending=False).head(10)
```
